```python
import matplotlib.pyplot as plt
import numpy as np

def plot_nanotube(m, n):
    # Chiral vector
    chiral_vector = np.array([m, n])

    # Create the unit vectors for the hexagonal lattice
    a1 = np.array([1, 0])
    a2 = np.array([1/2, np.sqrt(3)/2])

    # Generate the lattice points
    lattice_points = set()
    for i in range(-10, 11):
        for j in range(-10, 11):
            lattice_points.add(tuple(i*a1 + j*a2))

    # Plot the lattice points
    for point in lattice_points:
        plt.scatter(point[0], point[1], color='black', s=50)

    # Plot the nanotube vector
    plt.quiver(0, 0, chiral_vector[0], chiral_vector[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.008)

    # Set plot limits
    plt.xlim(-2, 10)
    plt.ylim(-2, 10)

    # Set aspect ratio to be equal
    plt.gca().set_aspect('equal', adjustable='box')

    # Add labels and title
    plt.xlabel('a1')
    plt.ylabel('a2')
    plt.title(f'Chirality ({m},{n}) Carbon Nanotube')

    # Show the plot
    plt.grid(True)
    plt.show()

# Example usage
plot_nanotube(6, 2)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

def plot_3d_nanotube(m, n, length=10, radius=1, num_points=100):
    theta = np.linspace(0, 2*np.pi, num_points)
    z = np.linspace(0, length, num_points)

    theta, z = np.meshgrid(theta, z)

    x = radius * np.cos(theta)
    y = radius * np.sin(theta)

    # Chiral vector
    chiral_vector = np.array([m, n])

    # Transformation matrix to rotate nanotube
    rotation_matrix = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])

    # Apply rotation to nanotube points
    nanotube_points = np.vstack((x.flatten(), y.flatten(), z.flatten()))
    nanotube_points = np.dot(rotation_matrix, nanotube_points)

    # Create 3D plot
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Plot the nanotube
    ax.plot_surface(nanotube_points[0].reshape(num_points, num_points),
                    nanotube_points[1].reshape(num_points, num_points),
                    nanotube_points[2].reshape(num_points, num_points),
                    color='red', alpha=0.7)

    # Set plot limits
    #ax.set_xlim([-length, length])
    #ax.set_ylim([-length, length])
    #ax.set_zlim([-length, length])

    # Add labels and title
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title(f'Chirality ({m},{n}) SWCNT')

    # Show the 3D plot
    plt.show()

# Example usage
plot_3d_nanotube(6, 2)
```
